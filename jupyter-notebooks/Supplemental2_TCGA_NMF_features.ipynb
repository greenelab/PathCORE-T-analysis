```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
tcga_nmf_file = "./data/tcga_data/NMF_model/NMF_300_features.tsv"
tcga_nmf_df = pd.read_table(tcga_nmf_file, sep="\t", index_col=0, header=None)

sns.set(color_codes=True)
```

```python
each_feature_is_nonnegative = (tcga_nmf_df >= 0).all()
print("The number of features in the NMF model is {0}".format(len(each_feature_is_nonnegative)))
print("All features are nonnegative: {0}".format(each_feature_is_nonnegative.all()))
```

```python
def plot_feature_weight_distributions(end_feature, dataframe):
    fig, ax = plt.subplots()
    # Features are numbered 1 through 300 because we index on column 0.
    for i in range(1, end_feature + 1):
        sns.distplot(dataframe.ix[:,i])
    ax.set_xlabel("Weight")
    ax.set_ylabel("Number of genes at weight")
    plt.show()
```

```python
# Plot the weight distribution of Feature #1
plot_feature_weight_distributions(1, tcga_nmf_df)
```

```python
# Plot the weight distribution of Features #1-30
plot_feature_weight_distributions(30, tcga_nmf_df)
```

```python
# Plot the weight distribution of Features #1-100
plot_feature_weight_distributions(100, tcga_nmf_df)
```

```python
# Plot the weight distribution of Features #1-300
plot_feature_weight_distributions(300, tcga_nmf_df)
```
